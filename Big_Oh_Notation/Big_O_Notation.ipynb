#### Time Complexity

* Big Oh is the language and metric we use to describe the efficiency of Algorithms.
* Time complexity is the way of showing of how the runtime of a function increases as the size of input increases.
* Big-Oh: It is the complexity that is going to be less or equal to the worst case.
* Big-Omega: It is the complexity that is going to be atleast more than the best case.
* Big-Theta: It is the complexity that is within the bounds of the worst and best cases.

#### Algorithm run time complexities

**Examples:**

|Complexity|Name|Sample|
|---|---|---|
|O(1)|Constant|Accessing a specific element in an array| 
|O(N)|Linear|Loop through array elements|
|O(Log(N))|Logarithmic|Find an element in sorted array|
|O(N2)|Quadratic|Looking at every index in an array twice|
|O(2N)|Exponential|Double recursion in Fibonacci|

#### Space Complexity

* Space complexity is a measure of amount of working storage an algorithm needs. 

#### Time complexity of recursive calls

```python

def findMaxNumRec(array, n):
    if n == 1:
        return array[0]
    else:
        return max(array[n-1], findMaxNumRec(array, n-1))
```


* The time complexity of the above function is O(N).

```python

def f3(n):
    if n <= 0:
        return 1
    else:
        return 1 + f3(n/5)
```

* The time complexity of above function is O(logn) as in recursive call input is getting divided.

***

* The **generalized** formuala to calculate the time complexity of the function is **_O(num of branches <sup>depth</sup>)_** if the number of recursive calls are more than 1.
* branches = number of branches in a recursive call tree.
* depth = Input param.

```python

def f(n):
    if n<=1:
        return 1
    else:
        return f(n-1) + f(n-1)
```

* num of branches in recursive call tree = 2
* depth = n
* Therefore, time complexity of above recursive function is O(2<sup>n</sup>).
