## Practical

```python
import numpy as np
```

```python
# Temperature readings 4 times a day, for 4 days

# Day 1 - 11, 15, 10, 6
# Day 2 - 10, 14, 11, 5
# Day 3 - 12, 17, 12, 8
# Day 4 - 15, 18, 14, 9
```

#### 1. Create 2D array

```python
two_d_array = np.array([
    [11, 15, 10, 6],
    [10, 14, 11, 5],
    [12, 17, 12, 8],
    [15, 18, 14, 9]
])

print(two_d_array)
```

#### 2. Insertion in 2D array

1. Using np.insert() function

```python
# Add column
col_add_two_d_arr = np.insert(two_d_array, 0, [[1,2,3,4]], axis=1) # for adding column axis=1 and for row axis=0
print(col_add_two_d_arr)
```

```python
# Add row
row_add_two_d_arr = np.insert(two_d_array, 1, [[1,2,3,4], [5,6,7,8]], axis=0)
print(row_add_two_d_arr)
```

2. Using np.append() function

```python
# Append row
row_appended_2_d_arr = np.append(two_d_array, [[1,2,3,4]], axis=0)
print(row_appended_2_d_arr)
```

```python
# Append column
col_appended_2_d_arr = np.append(two_d_array,[[1], [2], [3],[4]],axis=1)
print(col_appended_2_d_arr)
```

#### 3. Access element in 2D array

```python
def access_elements(arr, row_index, col_index):
    if row_index >= len(arr) or col_index >= len(arr[0]):
        print("Invalid dimensions!!")
    else:
        print(arr[row_index][col_index])

access_elements(two_d_array, 2, 2)
```

#### 4. Traverse 2D array

```python
def traverse_two_d_array(arr):
    for row_index in range(len(arr)):
        for col_index in range(len(arr[0])):
            print(arr[row_index][col_index])

traverse_two_d_array(two_d_array)
```

 #### 5. Search element in 2D array

```python
def search(arr, key):
    for row_index in range(len(arr)):
        for col_index in range(len(arr[0])):
            if arr[row_index][col_index] == key:
                return row_index, col_index
    return None

search_indexes = search(two_d_array, 14)

if search_indexes:
    print(f"Element found at index: {search_indexes[0]},{search_indexes[1]}")
else:
    print("Element not found")
```

####  6. Delete row or column from 2D array

```python
"""
np.delete(arr, index, axis)
row: axis = 0 
col: axis = 1
"""

# delete first row
arr = np.delete(two_d_array, 0, axis=0)
print(arr)

# delete first col
arr = np.delete(two_d_array, 0, axis=1)
print(arr)
```
