## Theory

***

####  Properties of Array

* Array can store elements of a specified data type only, which is used while declaring an array.
* Elements of an array are stored in contiguous locations in memory.
* Each element of an array has unique index(starts from 0).
* The size of array is predefined and cannot be modified.

#### Types of Array

* One-dimensional
* Multi-dimensional
* Regardless of the dimension of an array,all elements are stored continguosly in the memory.

## Practical

***

```python
import array
```

#### 1. Create an Array

```python
"""
Syntax:
arr = array.array(typecode, [initializers])

typecode denotes the datatype of the element of an array.
"""

arr1 = array.array('i', [1,2,3,4,5,6])
arr2 = array.array('d', [1.2,1.4,1.6])
print(arr1)
print(arr2)
```

#### 2. Insertion operation

```python
# array.insert(index, value)
arr1 = array.array('i', [1,2,3,4,5,6])
arr1.insert(6,7)
print("Insert element at the end: ", arr1)
```

```python
arr1 = array.array('i', [1,2,3,4,5,6])
arr1.insert(0,10)
print("Insert element at the beginning:", arr1)
```

#### 3. Traverse array

```python
arr1 = array.array('i', [1,2,3,4,5,6])
def traverse_array(arr):
    for el in arr:
        print(el)

traverse_array(arr1)
```

#### 4. Access array element

```python
arr1 = array.array('i', [1,2,3,4,5,6])

def access_element(arr, index):
    if index >= len(arr):
        return "Element at this index does not exist"
    else:
        return arr[index]

print(access_element(arr1, 6))
print(access_element(arr1, 1))
```

#### 5. Search an element in an array

```python
# arr.index(value) -> return the index of the first occurrence of value if it exists, otherwise raise ValueError 

arr1 = array.array("i", [1,2,3,4,5,6])
def search_array(arr, key):
    for el in arr:
        if el == key:
            return True
    return False

search_res = "Element found"  if search_array(arr1, 2) else "Element does not exist"
print(search_res)
```

#### 6. Delete element from array

```python
# arr.remove(value) -> remove the first occurrence of value from the array

arr1 = array.array('i')

for i in range(1,6):
    arr1.append(i)

print("Array before removing element 2: ", arr1)

arr1.remove(2)

print("Array after removing element 2: ", arr1)
```

#### 7.Pop(delete) an element from an array

```python
# arr.pop([i]) # by default return and delete element at last index, otherwise from index passed as parameter

arr1 = array.array('i')

for i in range(1,6):
    arr1.append(i)

# remove last element
last_element = arr1.pop()
print(last_element)

# remove first element
first_element = arr1.pop(0)
print(first_element)

print(arr1)
```
