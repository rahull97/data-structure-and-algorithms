```python
import array
```

### 1. Create an array and traverse

```python
arr = array.array('i', [1,2,3,4,5])

for el in arr:
    print(el)
```

### 2. Access individual elements through indexes

```python
print(arr[0])
print(arr[3])
```

### 3. Append any value to the array using append() method

```python
arr.append(6)
print(arr)
```

### 4. Insert value in an array using insert() method

```python
arr.insert(3, 11)
print(arr)
```

### 5. Extend array using extend() method

```python
arr1 = array.array('i', [10,11,12])
arr.extend(arr1)
print(arr)
```

### 6. Add items from list into array using fromlist() method

```python
arr.fromlist([20,21,22])
print(arr)
```

### 7. Remove any array element using remove() method

```python
arr.remove(11)
print(arr)
```

### 8. Remove last array element using pop() method 

```python
arr.pop()
print(arr)
```

### 9. Fetch any element's index using index() method

```python
print(arr.index(21))
```

### 10. Reverse an array using reverse() method

```python
arr.reverse()
print(arr)
```

### 11. Get array buffer information through buffer_info() method

```python
# provides array buffer start address in the memory, and the number of elements in the array
print(arr.buffer_info())
```

### 12. Check for number of occurrences of an element using count() method

```python
arr.append(11)
print(arr)
print(arr.count(11))
```

### 13. Convert array to its byte representation using tobytes() method

```python
bytes_representation = arr.tobytes()
print(bytes_representation)

arr_temp = array.array('i')
arr_temp.frombytes(bytes_representation)
print(arr_temp)
```
