# List

* It is a built-in mutable data structure in Python.
* It is a collection of an ordered elements
* It can store values having different data types

---

### 1. Create a List 

```python
li = [1,2,[10,12], "Hello"]
print(li)
```

### 2. Access elements and Traverse a List

```python
li = [10, 20, 30]

print(li[0])
print(li[-1])

print("\n")

for el in li:
    print(el)
    
print("\n")

for index in range(len(li)):
    li[index] = li[index] + 10

print(li)
```

### 3. Update and Insert elements in a List

```python
li = [1, 2, 3, 4]
li[0] = 10
print(li)

# insert element
li.insert(2, 30)
print(li)

# append element
li.append(100)
print(li)

# extend list
li.extend([101,102,103,104])
print(li)
```

### 4. Delete elements from List

```python
li = [10, 20, 30, 40, 50]

# pop()
el = li.pop() # delete last element
print(el)
print(li)

el = li.pop(0) # delete first element
print(el)
print(li)

print("\n")

# del operator
li = [10, 20, 30, 40, 50]
print(li)

del li[0:2]

print(li)

print("\n")

# remove method
li  = [10, 20, 30, 40, 50, 50]

li.remove(20)
li.remove(50)

print(li)
```

### 5. Search element in the List

```python
li  = [10, 20, 30, 40, 50, 60, 70, 80, 90]
key = 20

# using "in" keyword
if key in li:
    print(li.index(key))
else:
    print("key not found")
    

# using linear search

key = 40

def search(l, k):
    for el in l:
        if el == k:
            return l.index(k)
        
    return None

print(search(li, key))
```

### 6. List operations and functions

```python
# + operator
li1 = [10, 20, 30]
li2 = [40, 50, 60]
li3 = li1 + li2
print(li3, "\n")

# * operator
li = [10, 20]
li1 = li * 4
print(li1, "\n")

li = [10, 20, 30, 40, 50]

# len() function
print(len(li), "\n")

# max() and min() function
print(max(li))
print(min(li))
print("\n")

# sum() function
print(sum(li))
print(f"average: {sum(li)/len(li)}")
```

### 7. Lists and Strings

```python
string = "Hello World Hello World"

li = string.split(" ")
print(li)

string1 = ','.join(li)
print(string1)

string = "Hello"

li_of_chars = list(string)
print(li_of_chars)
```
